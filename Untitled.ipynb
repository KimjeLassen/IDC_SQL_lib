```python
import Main
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

binary = Main.get_binary_matrix()
clusters = Main.clusters(binary)
kmeans = Main.get_kmeans_labels(clusters, binary)
hierc = Main.get_hierarchical_labels(clusters, binary)
reduced = Main.get_reduced_data(binary)

Main.plot_clusters(reduced, kmeans, "K-Means Clustering")

# Hierarchical Clustering Plot
Main.plot_clusters(reduced, hierc, "Hierarchical Clustering")

# Step 6: Generate Dendrogram for Hierarchical Clustering
linked = linkage(binary, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', labels=binary.index.tolist(), distance_sort='descending', show_leaf_counts=False)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Users")
plt.ylabel("Euclidean Distance")
plt.show()
```
